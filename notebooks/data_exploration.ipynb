```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Define data paths
processed_data_path = r'C:\Users\tom.zbc\PycharmProjects\raman2Project\data\processed'
```

```python
# Load processed data
def load_data(sheet_name):
    file_path = os.path.join(processed_data_path, f'{sheet_name}.csv')
    return pd.read_csv(file_path)
```

```python
# Explore data
def explore_data(df):
    print("First few rows of the dataframe:")
    display(df.head())  # display() fungerer bedst i Jupyter Notebook

    print("\nData types and missing values:")
    df.info()  # Brug df.info() direkte, da det allerede printer output

    print("\nSummary statistics:")
    display(df.describe())

    print("\nMissing values in each column:")
    display(df.isnull().sum())

    # Plot correlation matrix
    plt.figure(figsize=(12, 8))
    sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
    plt.title("Correlation Matrix")
    plt.show()
```

```python
# Cleaning data
def clean_data(df):
    target_column = 'target'  # Juster dette, hvis din target-kolonne har et andet navn
    if target_column not in df.columns:
        print(f"Target column '{target_column}' not found. Please check your data.")
        return df

    df_cleaned = df.dropna(subset=[target_column])

    df_cleaned = df_cleaned.fillna(df_cleaned.median())

    print("\nData after cleaning:")
    df_cleaned.info()  # Brug df_cleaned.info() direkte

    return df_cleaned
```

```python
# Save cleaned data
def save_cleaned_data(df, sheet_name):
    cleaned_data_path = os.path.join(processed_data_path, f'{sheet_name}_cleaned.csv')
    df.to_csv(cleaned_data_path, index=False)
    print(f"Cleaned data saved to {cleaned_data_path}")
```

```python
# Main execution
sheets = ['CaF2', 'E_coli_25', 'E_coli_35', 'MRSE', 'MSSE', 'S_hom', 'S_lug']

for sheet in sheets:
    print(f"\nProcessing sheet: {sheet}")
    df = load_data(sheet)
    explore_data(df)
    df_cleaned = clean_data(df)
    if df_cleaned is not None:
        save_cleaned_data(df_cleaned, sheet)
```
